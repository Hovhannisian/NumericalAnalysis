TASK: find the minimum of $f(x)=x^4+x^2-2x $ on the  interval $[-4,5]$. Show  that $f(x)$ is unimodal on $[-4,5]$ and calculate the approximation of the minimum point using  the  golden section search method, with $\rho=\frac{3-\sqrt{5}}{2}$. Use as stopping condition $|b_n - a_n| < 10^{-3} $.

```python
import math
```

```python
def f(x):
    return x**4+x**2-2*x
a = -4
b = 5
ro =  (3- math.sqrt(5)) / 2
epsilion = 10**(-3)
```

```python
while b - a >= epsilion:
    A = a + ro * (b - a)
    B = b - ro * (b- a)
    
    if f(A) > f(B):
        a = A
    else:
        b = B
        
(a  + b) / 2   
```
