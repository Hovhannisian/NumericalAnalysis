Linear least square

```python
import matplotlib.pyplot as plt
from numpy import linspace, array, dot, polyfit, polyval, arange
```

```python
x = [-3, 1, 0, 7]
y = [2, 4, -1, 3]
```

```python
xx = linspace(0, 12, 1000)
coef = polyfit(x, y, 1)
yy = polyval(coef, xx)

```

```python
plt.figure(figsize=[9,8])
plt.xticks(arange(0,10))
plt.yticks(arange(0,10))
plt.axis([0, 10, 0, 10])
plt.grid(True)
plt.plot(x, y, "ro", label="Data")
plt.plot(xx, yy, label ="Least squares")

plt.legend()
```
